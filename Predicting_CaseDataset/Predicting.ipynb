# QnA

```python
# import time
import os
import pandas as pd
from transformers import pipeline
# os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID" 
# os.environ["CUDA_VISIBLE_DEVICES"] = "1"
```

```python
model_name = "deepset/xlm-roberta-large-squad2"
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name, device=0)
```

```python
questions = ['what does customer wants?', 'what does the customer needs help with?', "is customer's problem resolved?",
'does the brand resolved the problem?', "does the brand resolved the customer's problem?", 'what does customer asked for?', 'what solution does the brand provide?']
```

```python
file = '' # add file path here
qna = pd.read_csv(file)
```

```python
def pred(text, question):
    res = {}
    res['question'] = question
    res['context'] = text
    ans = nlp(res, max_answer_len=5)
    return ans['answer']
```

```python
def h(text):
    return pred(text, 'customer concerned about which product?').strip()
qna['Product'] = qna['Message'].apply(h)
qna
```

```python
qna
```

```python
def pred1(text, question):
    res = {}
    res['question'] = question
    res['context'] = text
    ans = nlp(res, max_answer_len=2)
    if ans['score'] < 5e-06:
        return 'NA'
    return ans['answer']
```

```python
def h(text):
    return pred1(text, 'what is the name of the brand involved?').strip()
qna['Brand'] = qna['Message'].apply(h)
qna
```

```python
def pred2(text, question):
    res = {}
    res['question'] = question
    res['context'] = text
    ans = nlp(res, max_answer_len=24)
    return ans['answer']
```

```python
def h(text):
    return pred2(text, 'what is the query of the customer?').strip()
qna['Customer_Query'] = qna['Message'].apply(h)
qna
```

```python
def h(text):
    return pred2(text, 'what is the problem of the customer').strip()
qna['Problem'] = qna['Message'].apply(h)
qna
```

```python
def h(text):
    return pred2(text, 'what is the action taken by executive?').strip()
qna['Brand_Reply'] = qna['Message'].apply(h)
qna
```

```python
qna.to_csv('QnA.csv', index=None) # add file path and name 
```

# Predicting Summary

```python
file_path = 'cleaned_data.csv' # add file path here
data = pd.read_csv(file_path)
data = data[:1000]
```

```python
model_name='vaibhavagg303/Bart_for_summarization_2'
summarizer = pipeline("summarization", model=model_name, tokenizer='facebook/bart-large-cnn', device=1)
```

```python
s = pd.DataFrame({'Message' : data['Message'], 'CaseNumber' : data['CaseNumber']})
```

```python
where = 1
def get(txt):
    global where
    if where % 100 == 0:
        print(where)
    where += 1
    return summarizer(txt)[0]['summary_text']
s['Summary'] = s['Message'].apply(get)
```

```python
s.to_csv('Bart2.csv', index=None)
```
