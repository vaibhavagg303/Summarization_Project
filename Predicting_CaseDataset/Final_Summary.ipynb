```python
import pandas as pd
from transformers import pipeline
```

```python
data = pd.read_csv('cleaned_data.csv')[:1000]
```

### Getting QnA

```python
model_name = "deepset/xlm-roberta-large-squad2"
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name, device=1)
```

```python
qna = data.copy()
```

```python
# Question 1 about product
def pred_product(text):
    question = 'customer concerned about which product?'
    res = {}
    res['question'] = question
    res['context'] = text
    ans = nlp(res, max_answer_len=5)
    return ans['answer'].strip()
qna['Product'] = qna['Message'].apply(pred_product)
```

```python
# Question 2 about brand
def pred_brand(text):
    question = 'what is the name of the brand involved?'
    res = {}
    res['question'] = question
    res['context'] = text
    ans = nlp(res, max_answer_len=2)
    if ans['score'] < 5e-06:
        return 'NA'
    return ans['answer'].strip()
qna['Brand'] = qna['Message'].apply(pred_brand)
```

```python
def pred_general(text, question):
    res = {}
    res['question'] = question
    res['context'] = text
    ans = nlp(res, max_answer_len=24)
    return ans['answer']
```

```python
# Question 3 about Customer's Query
def h(text):
    return pred_general(text, 'what is the query of the customer?').strip()
qna['Customer_Query'] = qna['Message'].apply(h)
```

```python
# Question 4 about Customer's Problem
def h(text):
    return pred_general(text, 'what is the problem of the customer').strip()
qna['Problem'] = qna['Message'].apply(h)
```

```python
# Question 5 about Brand's reply
def h(text):
    return pred_general(text, 'what is the action taken by agent?').strip()
qna['Brand_Reply'] = qna['Message'].apply(h)
```

```python
# Question 6 about Brand's asks
def h(text):
    return pred_general(text, 'what does agent asks for?').strip()
qna['Brand_Asks'] = qna['Message'].apply(h)
```

### Getting short summary

```python
pegasus = data.copy()
```

```python
model_name='vaibhavagg303/Pegasus-Large' # model_name import from hugging face
summarizer = pipeline("summarization", model=model_name, tokenizer='google/pegasus-large', device=1)
```

```python
def get(txt):
    return summarizer(txt)[0]['summary_text']
pegasus['Summary'] = pegasus['Message'].apply(get)
```

```python
def check(msg):
    if 'agent' in msg:
        return True
    return False
pegasus['brand'] = pegasus['Message'].apply(check)
```

```python
qna
```

```python
def update(row):
    msg = row['Summary']
    length = len(msg.split('. '))
    if row['brand'] == False:
        msg = 'customer ' + msg
    elif length >= 2:
        msg = msg.split('. ')
        msg[0] = 'customer ' + msg[0]
        msg[-1] = 'agent ' + msg[-1]
        msg = '. '.join(msg)
    elif msg.split(' ')[0] == 'tells':
        msg = 'agent ' + msg
    elif msg.split(' ')[0] != 'asks':
        msg = 'customer ' + msg
    else:
        if 'survey' in msg:
            msg = 'agent ' + msg
        elif 'help' in msg or 'about' in msg or 'for' in msg:
            msg = 'customer ' + msg
        else: msg = 'agent ' + msg
    return msg
pegasus['new_s'] = pegasus.apply(update, axis=1)
```

```python
for i in range(1000):
    msg = pegasus['Summary'][i].split('. ')
    if msg[0].split(' ')[0] == 'tells' and len(msg) == 1:
        print(i, msg)
```

### Final Summary

```python
final = data.copy()
```

```python
final_sum = []
```

```python
for i in range(1000):
    summary = pegasus['new_s'][i]
    product = qna['Product'][i]
    if 'customer' not in summary:
        summary = 'customer says ' + qna['Customer_Query'][i] + '. ' + summary
    if product not in summary:
        summary = 'customer needs help with ' + product + '. ' + summary
    if ('agent' not in summary) and pegasus['brand'][i] == True:
        add = qna['Brand_Reply'][i]
        summary += ' agent replied ' + add + '.'
    if pegasus['brand'][i] == False:
        summary += '. customer is still waiting for help.'    
    summary = summary.replace('..', '.')
    final_sum.append(summary)
final['Summary'] = final_sum
```

```python
final.to_csv('final_summary.csv', index=None)
```
