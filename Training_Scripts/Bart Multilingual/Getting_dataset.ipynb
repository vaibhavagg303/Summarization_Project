```python
import pandas as pd
from datasets import load_dataset
```

```python
def get_data(language, len_train, len_val, len_test):
    dataset = load_dataset('mlsum', language)
    train = pd.DataFrame({'text': dataset['train']['text'], 'summary': dataset['train']['summary']})[:len_train]
    test = pd.DataFrame({'text': dataset['test']['text'], 'summary': dataset['test']['summary']})[:len_test]
    val = pd.DataFrame({'text': dataset['validation']['text'], 'summary': dataset['validation']['summary']})[:len_val]
    return train, test, val
```

```python
train_german, test_german, val_german = get_data('de', 25000, 1000, 1000)
```

```python
train_spanish, test_spanish, val_spanish = get_data('es', 25000, 1000, 1000)
```

```python
train_french, test_french, val_french = get_data('fr', 25000, 1000, 1000)
```

```python
train_russian, test_russian, val_russian = get_data('ru', 25000, 1000, 1000)
```

```python
train_turkish, test_turkish, val_turkish = get_data('tu', 25000, 1000, 1000)
```

```python
train_data = pd.concat([train_german, train_spanish, train_russian, train_french, train_turkish], ignore_index=True)
```

```python
test_data = pd.concat([test_german, test_spanish, test_russian, test_french, test_turkish], ignore_index=True)
```

```python
val_data = pd.concat([val_german, val_spanish, val_russian, val_french, val_turkish], ignore_index=True)
```

```python
train_data.to_csv('Dataset/train.csv', index=None)
```
