```python
import pandas as pd
```

```python
train_df = pd.read_csv('DialogSum_dataset/train.csv')
test_df = pd.read_csv('DialogSum_Dataset/test.csv')
val_df = pd.read_csv('DialogSum_Dataset/val.csv')
```

```python
train_df['text'] = train_df['dialogue']
test_df['text'] = test_df['dialogue']
val_df['text'] = val_df['dialogue']
test_df['summary'] = test_df['summary2']
```

```python
train_df.drop(['dialogue', 'topic'], inplace=True, axis=1)
val_df.drop(['dialogue', 'topic'], inplace=True, axis=1)
test_df.drop(['topic1', 'topic2', 'topic3', 'summary1', 'summary2', 'summary3', 'dialogue'], axis=1, inplace=True)
```

```python
def clean(msg):
    msg = msg.replace('\n', ' ')
    msg = msg.replace('#', '')
    msg = msg.replace('\r', '')
    msg = msg.replace('Person1:', 'customer asked')
    msg = msg.replace('Person2:', 'agent replied')
    return msg
train_df['text'] = train_df['text'].apply(clean)
train_df['summary'] = train_df['summary'].apply(clean)
test_df['text'] = test_df['text'].apply(clean)
test_df['summary'] = test_df['summary'].apply(clean)
val_df['text'] = val_df['text'].apply(clean)
val_df['summary'] = val_df['summary'].apply(clean)
```

```python
import re
def clean_data(msg):
    msg = re.sub('(www[^\s]*) | (https?[^\s]*)', '', msg) # removing links
    msg = re.sub('([^\s]*@[^\s]+)', '', msg)              # removing usernames, emails
    msg = re.sub(r'(\S*)[\d]+[^\s]*', '', msg)            # removing numbers
    msg = re.sub(r'\s+', ' ', msg)                        # removing extra spaces
    msg = re.sub('[\n]', ' ', msg)                        # removing new lines
    return msg.encode('ascii', 'ignore').decode('ascii')  # returning demojified text
train_df['text'] = train_df['text'].apply(clean_data)
train_df['summary'] = train_df['summary'].apply(clean_data)
test_df['text'] = test_df['text'].apply(clean_data)
test_df['summary'] = test_df['summary'].apply(clean_data)
val_df['text'] = val_df['text'].apply(clean_data)
val_df['summary'] = val_df['summary'].apply(clean_data)
```

```python
train_df
```

```python
# swapping columns text and summary
columns_titles = ["text","summary"]
train_df = train_df.reindex(columns=columns_titles)
val_df = val_df.reindex(columns=columns_titles)
train_df
```

```python
# adding some context to the conversation
def add_context(msg):
    msg = 'this is the conversation between customer and agent. ' + msg
    return msg
train_df['text'] = train_df['text'].apply(add_context)
test_df['text'] = test_df['text'].apply(add_context)
val_df['text'] = val_df['text'].apply(add_context)
```

```python
# saving final files in Cleaned_dataset folder
train_df.to_csv('Cleaned_dataset/train.csv', index=None)
test_df.to_csv('Cleaned_dataset/test.csv', index=None)
val_df.to_csv('Cleaned_dataset/val.csv', index=None)
```
